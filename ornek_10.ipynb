ÖRNEK 10: Verilen bir sayıya kadar verilen bir sayının kalan sınıflarını bulalım.

```python
def kalan_siniflari(ust_sinir, mod):
    sonuc = {i: [] for i in range(mod)}
    for i in range(ust_sinir + 1):
        sonuc[i % mod].append(i)
    return sonuc

# Kullanıcıdan giriş almak
ust_sinir = int(input("Üst sınırı girin: "))
mod = int(input("Mod değerini girin: "))

# Sonucu hesaplayıp ekrana yazdırma
sonuclar = kalan_siniflari(ust_sinir, mod)
for kalan, sayilar in sonuclar.items():
    print(f"{kalan} mod {mod} => {sayilar}")
```

```python
from tabulate import tabulate

def kalan_siniflari(ust_sinir, mod):
    sonuc = {i: [] for i in range(mod)}
    for i in range(ust_sinir + 1):
        sonuc[i % mod].append(i)
    return sonuc

# Kullanıcıdan giriş almak
ust_sinir = int(input("Üst sınırı girin: "))
mod = int(input("Mod değerini girin: "))

# Sonucu hesaplayıp tablo olarak ekrana yazdırma
sonuclar = kalan_siniflari(ust_sinir, mod)
tablo = [[kalan, ", ".join(map(str, sayilar))] for kalan, sayilar in sonuclar.items()]
print(tabulate(tablo, headers=["Kalan", "Sayilar"], tablefmt="grid"))
```
